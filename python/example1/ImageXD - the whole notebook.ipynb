```python
def normalize_rectangle(rect):
    """
    Normalizes a rectangle so that it is at the origina and 1.0 units 
    Parameters
    ----------
    rect : list
        coordinates
    
    
    """
    assert len(rect)==4, "Input must have 4 parameters"
    x0,y0,x1,y1 = rect
    assert x0<x1, "x0 must be smaller than x1"
    assert y0<y1, "y0 must be smaller than y1"
    
    dx = x1-x0
    dy = y1 -y0
    if dx>dy:
        scaled = float(dy)/dx
        upper_x,upper_y = 1.0, scaled
    else:
        scaled = float(dx)/dy
        upper_x, upper_y = scaled, 1.0
    
    #does your answer makes sense??? use assert as well
    assert upper_x > 0 and upper_x <=1.0, "Calculated upper"
    
    return (0,0, upper_x, upper_y)
            
```
